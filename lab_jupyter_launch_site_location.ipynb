```python
# Task 2: Mark the success/failed launches for each site on the map
site_map = folium.Map(location=[launch_sites_df['Lat'].mean(), launch_sites_df['Long'].mean()], zoom_start=5)

# Add circles and markers for each launch site
for _, site in launch_sites_df.iterrows():
    coordinate = [site['Lat'], site['Long']]
    site_name = site['Launch Site']
    
    # Add a circle at the launch site
    circle = folium.Circle(
        coordinate, 
        radius=1000, 
        color='#d35400', 
        fill=True, 
        fill_color='#d35400'
    ).add_child(folium.Popup(site_name))
    
    # Add a custom marker for the site
    marker = folium.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:#d35400;"><b>{site_name}</b></div>'
        )
    )
    
    # Add the circle and marker to the map
    site_map.add_child(circle)
    site_map.add_child(marker)

# Now, add markers for the launch results (success/failure)
for _, launch in spacex_df.iterrows():
    coordinate = [launch['Lat'], launch['Long']]
    site_name = launch['Launch Site']
    launch_class = launch['class']
    
    # Color the marker based on success or failure (Green for success, Red for failure)
    if launch_class == 1:  # Success
        color = 'green'
        message = 'Success'
    else:  # Failure
        color = 'red'
        message = 'Failure'
    
    # Add a circle for the launch result
    result_circle = folium.Circle(
        coordinate, 
        radius=500,  # Smaller radius for launch result markers
        color=color, 
        fill=True, 
        fill_color=color
    ).add_child(folium.Popup(f'{site_name}: {message}'))
    
    # Add a marker for the launch result
    result_marker = folium.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12; color:{color};"><b>{message}</b></div>'
        )
    )
    
    # Add the result circle and marker to the map
    site_map.add_child(result_circle)
    site_map.add_child(result_marker)

# Display the map
site_map
```

```python

# TASK 3: Calculate the distances between a launch site to its proximities
from geopy.distance import geodesic
def calculate_distance(lat1, lon1, lat2, lon2):
    return geodesic((lat1, lon1), (lat2, lon2)).kilometers

# Create an empty dictionary to store the distances between launch sites
distance_dict = {}

# Loop through each pair of launch sites and calculate the distance between them
for i, row1 in launch_sites_df.iterrows():
    for j, row2 in launch_sites_df.iterrows():
        if i != j:  # Exclude the pair where both are the same site
            # Calculate the distance between launch site i and launch site j
            distance = calculate_distance(row1['Lat'], row1['Long'], row2['Lat'], row2['Long'])
            distance_dict[(row1['Launch Site'], row2['Launch Site'])] = distance

# Convert the dictionary to a DataFrame for better readability
distance_df = pd.DataFrame(list(distance_dict.items()), columns=['Launch Sites', 'Distance_km'])

# Display the resulting distance DataFrame
distance_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
